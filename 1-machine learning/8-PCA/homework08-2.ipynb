```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt

mat=sio.loadmat('ex7faces.mat')
X=mat['X']
X.shape
```

```python
def plot_100_images(X):
    fig,axs=plt.subplots(ncols=10,nrows=10,figsize=(10,10))
    for i in range(10):
        for j in range(10):
            axs[i,j].imshow(X[10*i+j].reshape(32,32).T,cmap='Greys_r')
            axs[i,j].set_xticks([])
            axs[i,j].set_yticks([])
```

```python
plot_100_images(X)
```

```python
means=np.mean(X,axis=0)
X_demean=X-means
C=X_demean.T@X_demean
U,S,V=np.linalg.svd(C)
```

```python
U1=U[:,:36]
X_reductiom=X_demean@U1
X_reductiom.shape
```

```python
X_recover=X_reductiom@U1.T+means
```

```python
plot_100_images(X_recover),plot_100_images(X)
```
